```python
conjunto_paises = {'España', 'Francia', 'Italia', 'Luxemburgo', 'Argentina', 'Colombia'}
print(conjunto_paises)
```

```python
paises_norteamerica = ('Mexico', 'Canada', 'Estados Unidos')
print(paises_norteamerica)
```

```python
conjunto_paises.update(paises_norteamerica)
print(conjunto_paises)
```

```python
conjunto_paises.remove('Italia')
print(conjunto_paises)
```

```python
paises_asia = {'China', 'Japon', 'Korea'}
```

```python
paises_mundial = conjunto_paises.union(paises_asia)
```

```python
print(paises_mundial)
```

```python
paises2 = paises_mundial.intersection(paises_norteamerica)
print(paises2)
```

```python
paises3 = paises_mundial.difference(paises_norteamerica)
print(paises3)
```
