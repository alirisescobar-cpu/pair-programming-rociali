```python
def es_palindromo(unacadena):
    if unacadena == unacadena[::-1]:
        print(unacadena, 'es un palindromo')
    else:
        print(unacadena, 'no es un palindromo')
```

Descripción: Crea una función llamada calcular_promedio que tome una lista de números como parámetro y devuelva el promedio de todos sus elementos.

```python
def calcular_promedio(una_lista):
    promedio = (sum(una_lista)) / (len(una_lista))
    return promedio
```

```python
lista55 = [55, 46, 38, 90]
calcular_promedio(lista55)
```
